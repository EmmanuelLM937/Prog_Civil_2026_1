# Problema de ingeniería (Nivel básico)

Supongamos un problema típico en **Ingeniería Civil o Agrícola**:

> Calcular el **volumen de agua almacenado en un tanque rectangular**.

Este tipo de cálculo se usa en:

* almacenamiento de agua para riego
* reservorios agrícolas
* tanques de sistemas hidráulicos

---

# 1. Definir el algoritmo (Entradas – Proceso – Salidas)

### Entradas (datos que proporciona el usuario)

* `largo` del tanque (m)
* `ancho` del tanque (m)
* `altura` del tanque (m)

### Proceso (operación matemática)

El volumen de un prisma rectangular es:

[
Volumen = largo \times ancho \times altura
]

### Salida

* `volumen` del tanque en **metros cúbicos (m³)**

---

# 2. Diseño del algoritmo paso a paso

1. Iniciar el programa
2. Leer `largo`
3. Leer `ancho`
4. Leer `altura`
5. Calcular `volumen = largo × ancho × altura`
6. Mostrar el volumen
7. Terminar

---

# 3. Pseudocódigo

```
Inicio

    Escribir "Ingrese el largo del tanque (m):"
    Leer largo

    Escribir "Ingrese el ancho del tanque (m):"
    Leer ancho

    Escribir "Ingrese la altura del tanque (m):"
    Leer altura

    volumen = largo * ancho * altura

    Escribir "El volumen del tanque es:", volumen, "m3"

Fin
```

---

# 4. Código en Python

```python
# Programa para calcular el volumen de un tanque rectangular
# Aplicación básica de ingeniería

# Entrada de datos
largo = float(input("Ingrese el largo del tanque (m): "))
ancho = float(input("Ingrese el ancho del tanque (m): "))
altura = float(input("Ingrese la altura del tanque (m): "))

# Proceso: cálculo del volumen
volumen = largo * ancho * altura

# Salida
print("El volumen del tanque es:", volumen, "m3")
```

### Un ejemplo aplicado a la Ingeniería

Antes de mostrarte el código, pensemos en un problema común: **El cálculo del área de un terreno trapezoidal** (muy frecuente en topografía o diseño de canales).

En Python, no solo queremos el resultado, sino que el código sea "leíble" para otros colegas. Aquí tienes una estructura básica que podrías ejecutar en **Google Colab**:

````python
# Definición de una función para modularizar el cálculo
def calcular_area_trapecio(base_mayor, base_menor, altura):
    """
    Calcula el área de un trapecio dadas sus dimensiones.
    """
    area = ((base_mayor + base_menor) * altura) / 2
    return area

# Variables con nombres descriptivos (unidades incluidas)
b_mayor_m = 15.5
b_menor_m = 10.2
h_m = 5.0

# Llamada a la función y almacenamiento del resultado
resultado_area = calcular_area_trapecio(b_mayor_m, b_menor_m, h_m)

# Salida de datos
print(f"El área del terreno es: {resultado_area} metros cuadrados.")
````
