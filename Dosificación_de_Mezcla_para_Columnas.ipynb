# **Dosificación de Mezclas para Columnas**

 **Planteamiento:**

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

Se sabe que para un concreto estándar de 210 kg/cm2
(3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.


**Requerimientos**:

1. Solicitar la altura de la columna en metros.

2. Solicitar el diámetro de la columna en metros.

3. Calcular el volumen de la columna: $$v = \pi * r^2 * h$$


. (Sugerencia: usar 3.1416
 o importar math.pi).

4. Calcular la cantidad de sacos:
$$sacos = v * 9.5$$


**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2):


Volumen ≈0.377 m3


Sacos necesarios ≈3.58


 (el estudiante verá aquí la importancia de los tipos decimales).

````python
 # ENTRADA: Solicitar la altura de la columna en metros.
 # Solicitar el diámetro de la columna en metros.
 h = float(input("Ingrese la altura de la columna en metros: "))
 d = float(input("Ingrese el diámetro de la columna en metros: "))

 # PROCESO: Calcula el radio r, calcua el volumen de la columna, calcula sacos
 r = d/2
 v = 3.1416 * r**2 * h
 sacos = v * 9.5

 # SALIDA: Imprimir el volumen y la cantidad de sacos
 print (f"El volumen de la columna es: {v:.2f} m3")
 print (f"La cantidad de sacos necesarios es: {sacos:.2f}")
````

```markdown
El código para calcular los sacos de cemento funciona de la siguiente manera:

*   `h = float(input("Ingrese la altura de la columna en metros: "))`
    *   **Función**: Solicita al usuario que ingrese la altura de la columna en metros.
    *   **`input()`**: Muestra el mensaje entre paréntesis y espera a que el usuario escriba algo y presione Enter.
    *   **`float()`**: Convierte la entrada del usuario (que es siempre una cadena de texto) a un número decimal (flotante) para poder realizar cálculos matemáticos.
    *   **`h = ...`**: Asigna el valor numérico ingresado por el usuario a la variable `h`.

*   `d = float(input("Ingrese el diámetro de la columna en metros: "))`
    *   **Función**: De manera similar a la línea anterior, solicita al usuario el diámetro de la columna en metros y lo guarda como un número decimal en la variable `d`.

*   `r = d/2`
    *   **Función**: Calcula el radio de la columna.
    *   **`d/2`**: Divide el diámetro (`d`) entre 2 para obtener el radio.
    *   **`r = ...`**: Asigna el resultado de esta operación a la variable `r`.

*   `v = 3.1416 * r**2 * h`
    *   **Función**: Calcula el volumen de la columna circular.
    *   **`3.1416`**: Representa el valor aproximado de Pi (π).
    *   **`r**2`**: Calcula el radio (`r`) elevado al cuadrado (r * r).
    *   **`*`**: Multiplica el valor de Pi por el radio al cuadrado y luego por la altura (`h`).
    *   **`v = ...`**: Asigna el volumen calculado a la variable `v`.

*   `sacos = v * 9.5`
    *   **Función**: Calcula la cantidad de sacos de cemento necesarios.
    *   **`v * 9.5`**: Multiplica el volumen de la columna (`v`) por 9.5, que es la cantidad de sacos de cemento requeridos por cada metro cúbico de mezcla.
    *   **`sacos = ...`**: Asigna la cantidad total de sacos a la variable `sacos`.

*   `print (f"El volumen de la columna es: {v:.2f} m3")`
    *   **Función**: Muestra en la consola el volumen calculado de la columna.
    *   **`print()`**: Es una función para mostrar texto en la salida.
    *   **`f" ... {variable:.2f} ... "`**: Es una f-string (cadena formateada). Permite incrustar variables y expresiones dentro de cadenas de texto. El `: .2f` formatea la variable `v` para que se muestre con solo dos decimales.

*   `print (f"La cantidad de sacos necesarios es: {sacos:.2f}")`
    *   **Función**: Muestra en la consola la cantidad de sacos de cemento necesarios.
    *   Similar a la línea anterior, utiliza una f-string para mostrar el valor de la variable `sacos` formateado a dos decimales.
```
